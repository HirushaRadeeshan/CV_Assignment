```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('runway.png', cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(6,4))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

(a) Gamma correction with γ = 0.5.

```python
#Gamma correction with gamma = 0.5
gamma = 0.5
gamma_table = np.array([(i / 255.0) ** gamma * 255 for i in range(256)]).astype(np.uint8)
img_gamma_05 = cv2.LUT(img, gamma_table)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(img_gamma_05, cmap='gray')
plt.title(f'Gamma = {gamma}')
plt.axis('off')
plt.show()
```

(b) Gamma correction with γ = 2.

```python
# Gamma correction with gamma = 2
gamma = 2
gamma_table = np.array([(i / 255.0) ** gamma * 255 for i in range(256)]).astype(np.uint8)
img_gamma_2 = cv2.LUT(img, gamma_table)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(img_gamma_2, cmap='gray')
plt.title(f'Gamma = {gamma}')
plt.axis('off')
plt.show()
```

(c) Contrast Stretching (linear piecewise transformation)

```python
# Contrast stretching with r1 = 0.2, r2 = 0.8
r1, r2 = 0.2, 0.8

img_norm = img / 255.0
img_stretched = np.zeros_like(img_norm)
img_stretched[img_norm < r1] = 0
mask = (img_norm >= r1) & (img_norm <= r2)
img_stretched[mask] = (img_norm[mask] - r1) / (r2 - r1)
img_stretched[img_norm > r2] = 1
img_stretched = (img_stretched * 255).astype(np.uint8)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(img_stretched, cmap='gray')
plt.title('Contrast Stretched (r1=0.2, r2=0.8)')
plt.axis('off')
plt.show()
```
